### Reading a file :

```python
def read_data(filename):
    with open(filename,'r') as f:
        text = f.readlines()
        print(text)
        return text
```

```python
lines=read_data('welcome.txt')
```

## Each line separatley

```python
##not running a for lopp and converting the lines to a data frame
df_lines=pd.DataFrame(lines)
print("Printing Each line separately using DataFrame")
df_lines[0]
```

###  Number of Vowels

```python
paragraphs = df_lines[0]
count=0
vowel=['a','e','i','o','u',]
for line in paragraphs:
    for char in line:
        if(char in vowel):
            count=count+1
print(f"Count of Vowels:{count}")
```

### Frequency of each word

```python
word_count_dict={}
word=""
begin=0
for line in paragraphs:
    for char in line:
        word+=char
        if(char==' '):   
            if word:
                if(word in word_count_dict):
                    word_count_dict[word]+=1
                else:
                    word_count_dict[word]=1
                word=""
    if word:
        if word in word_count_dict:
            word_count_dict[word] += 1
        else:
            word_count_dict[word] = 1

for word, count in word_count_dict.items():
    print(f"'{word}': {count} times")
```

```python
data= data.iloc[:, :-1] 
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
```

```python
pca = PCA(n_components=2) 
pca_result = pca.fit_transform(scaled_data)
pca_df = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2'], index=data.index)
pca_result
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(pca_df['PC1'], pca_df['PC2'])

for i in pca_df.index:
    plt.text(pca_df['PC1'][i], pca_df['PC2'][i], str(i))

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Petal Characteristics ')
plt.show()

print(pca_df)
```

### Plots and visualizations

```python
import seaborn as sb
import matplotlib
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2,figsize=(8,10))
axs[0].set_title('Petals-Sepals in Iris')
axs[0].set_xlabel('Sepal length in cm');
axs[0].set_ylabel('Petal length in cm ');
sb.scatterplot(x=data['sepal.length'],y=data['petal.length'],hue=data.variety,ax=axs[0]);

axs[1].set_title('Box plot for petal Length across species')
sb.boxplot(x="variety", y="petal.length", palette="husl", data=data,ax=axs[1])
fig.tight_layout()
```

```python
sb.histplot(data=data, x="petal.length")
plt.title('Histogram');
```
