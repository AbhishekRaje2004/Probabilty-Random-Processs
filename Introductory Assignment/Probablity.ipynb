```python
sample_space=[1,2,3,4,5,6,7]
```

```python
event_A=[1,2]
event_B=[2,3,4]
```

```python
def sample(sample_space,event):
    count=0
    for element in sample_space:
        if(element in event):
            count=count+1
    return count
    
```

```python
sample(sample_space,event)/len(sample_space)
```

```python

joint_prob=len([x for x in event_A if x in event_B ])/len(sample_space)
```

```python
joint_prob
```

```python
#A/B
cond_prob=joint_prob/((len(event_B)/len(sample_space)))
```

```python
cond_prob
```

```python
prob=[0.2,0.7,0.1]
outcome=[10,20,30]
```

```python
def mean(prob,outcome):
    exp_val=0
    for value in prob:
        exp_val=exp_val+value*outcome[prob.index(value)]
    return exp_val
```

```python
mean(prob,outcome)
```
