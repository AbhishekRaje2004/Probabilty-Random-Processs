```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

### Generating normal distribution datset

```python
def generate_normal_distribution(mean=0, std_dev=1, num_samples=10000):
    data = np.random.normal(loc=mean, scale=std_dev, size=num_samples)
    return data
```

```python
def uniformly_sample_data(data, num_samples=1000):#taking thouand of them randomly
    sampled_indices = np.random.choice(len(data), size=num_samples, replace=False)
    sampled_data = data[sampled_indices]
    return sampled_data
```

```python
def reconstruct_distribution(sampled_data):
    mean, std_dev = norm.fit(sampled_data)
    return mean, std_dev
```

```python
def plot_distributions(original_data, sampled_data, reconstructed_mean, reconstructed_std_dev):
    plt.hist(original_data, bins=50, color='blue', density=True, label='Original Data')
    plt.hist(sampled_data, bins=30, color='green', density=True, label='Sampled Data')
    
    x = np.linspace(min(original_data), max(original_data), 1000)
    reconstructed_pdf = norm.pdf(x, loc=reconstructed_mean, scale=reconstructed_std_dev)
    
    plt.plot(x, reconstructed_pdf, 'r-', lw=2, label='Reconstructed Distribution')
    plt.legend()
    plt.title('Original vs Sampled vs Reconstructed Distributions')
    plt.xlabel('Data Values')
    plt.ylabel("Density")
    plt.show()
```

```python
mean, std_dev = 0, 1  
original_data = generate_normal_distribution(mean=mean, std_dev=std_dev, num_samples=10000)
```

```python
sampled_data = uniformly_sample_data(original_data, num_samples=1000)
```

```python
reconstructed_mean, reconstructed_std_dev = reconstruct_distribution(sampled_data)
```

```python
plot_distributions(original_data, sampled_data, reconstructed_mean, reconstructed_std_dev)
```
